```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('test.csv')
```

```python
def plot_confusion_matrix(y_true, y_pred, title):
    # Generate the confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Rearrange the confusion matrix (TP, FP, FN, TN)
    cm_rearranged = np.array([[cm[1, 1], cm[0, 1]], 
                              [cm[1, 0], cm[0, 0]]])

    # Plot the rearranged confusion matrix
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm_rearranged, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=['Actual Positive', 'Actual Negative'],
                yticklabels=['Predicted Positive', 'Predicted Negative'])

    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix: {title}')
    plt.show()
```

```python
# %% Confusion Matrix: label vs rule
y_true = df['label'].apply(lambda x: 1 if x != 0 else 0)

y_pred_rule = df['rule'].apply(lambda x: 1 if x != 0 else 0)
plot_confusion_matrix(y_true, y_pred_rule, 'Rulebased')

# %% Confusion Matrix: label vs ml
y_pred_ml = df['ml'].apply(lambda x: 1 if x != 0 else 0)
plot_confusion_matrix(y_true, y_pred_ml, 'ML classifier')

# %% Confusion Matrix: label vs llm
y_pred_llm = df['llm'].apply(lambda x: 1 if x != 0 else 0)
plot_confusion_matrix(y_true, y_pred_llm, 'LLM approach')

# %% Confusion Matrix: label vs combined
y_pred_combined = df['combined'].apply(lambda x: 1 if x != 0 else 0)
plot_confusion_matrix(y_true, y_pred_combined, 'Combined')
```
